### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import cluster
from sklearn.metrics import silhouette_score
```

### Read Dataset

```python
data = pd.read_csv('https://raw.githubusercontent.com/Rietaros/kampus_merdeka/main/cluster_s1.csv')
data = data.drop('no', axis = 1)
data.head()
```

```python
%matplotlib inline
ax1 = data.plot.scatter(x='x', y='y', c='DarkBlue')
```

### Data Segmentation (Clustering)

```python
# TODO: Lakukan pencarian jumlah cluster terbaik berdasarkan data diatas
best_k = 0
best_score = 0
K = range(2, 22)

for k in K:
    kmeans = cluster.KMeans(n_clusters=k, random_state=0)
    kmeans.fit(data)
    score = silhouette_score(data, kmeans.labels_)
    print(f"Silhouette Score for k={k}: {score:.5f}")

    if score > best_score:
        best_score = score
        best_k = k

print(f"Best k: {best_k} with Silhouette Score: {best_score:.5f}")
```

```python
# TODO: Lakukan pemodelan dengan KMeans
kmeans = cluster.KMeans(n_clusters=best_k, random_state=0)
kmeans.fit(data)
label = kmeans.labels_

score = silhouette_score(data, label)
print("Silhouette score={:.5f}".format(score))
```

### Visualization: Cluster Result

```python
# TODO: Masukkan label hasil clustering ke dalam DataFrame
result = data.copy()
result['cluster'] = label
result.head()
```

```python
# TODO: Plot data hasil clustering dengan Seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(x='x', y='y', hue='cluster', data=result, palette='Set1', s=100, alpha=0.7)
plt.title('KMeans Clustering Result')
plt.show()
```
